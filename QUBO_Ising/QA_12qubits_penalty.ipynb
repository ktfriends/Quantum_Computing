```python
# A = {{1, -1, 1}, {-2, 3, 4}, {1, -4, -1}}
# x1 = 2 q0 + q1 - 2 q2 - q3
# b = {4, -4, 8}
# 12 qubits with integer solutions
# QUBO minimum -96
# Ising minimum -446
# panelty terms are zero

from dwave.system import DWaveSampler, EmbeddingComposite

sampler_auto = EmbeddingComposite(DWaveSampler(solver={'qpu': True}))

linear = {('q0','q0'):-56, ('q1','q1'):-34 , ('q2','q2'):104, ('q3','q3'):46, 
          ('q4','q4'):296, ('q5','q5'):122, ('q6','q6'):-88, ('q7','q7'):-70,
         ('q8','q8'):152, ('q9','q9'):58, ('q10','q10'): -8, ('q11','q11'):-22}


quadratic = {('q0','q1'):24, ('q0','q2'):0, ('q1','q2'):0, ('q0','q3'):0,
             ('q1','q3'):0, ('q2','q3'):24, ('q0','q4'): - 88, ('q1','q4'):- 44, 
             ('q2','q4'):88,('q3','q4'):44, ('q0','q5'):-44,('q1','q5'):- 22, ('q2','q5'):44,
             ('q3','q5'):22, ('q4','q5'):104,('q0','q6'):88, ('q1','q6'):44,('q2','q6'):- 88, 
             ('q3','q6'):- 44,('q4','q6'):0, ('q5','q6'):0,('q0','q7'):44, ('q1','q7'):22,
             ('q2','q7'):- 44, ('q3','q7'):- 22,('q4','q7'):0, ('q5','q7'):0,('q6','q7'):104,
             ('q0','q8'):- 64, ('q1','q8'):- 32, ('q2','q8'):64, ('q3','q8'):32,
             ('q4','q8'):120, ('q5','q8'):60, ('q6','q8'):- 120, ('q7','q8'):- 60,
             ('q0','q9'):-32, ('q1','q9'):- 16, ('q2','q9'):32, ('q3','q9'):16,
             ('q4','q9'):60, ('q5','q9'):30, ('q6','q9'):- 60, ('q7','q9'):- 30, ('q8','q9'):72,  
             ('q0','q10'):64, ('q1','q10'):32, ('q2','q10'):- 64, ('q3','q10'):- 32,
             ('q4','q10'):- 120, ('q5','q10'):-60, ('q6','q10'):120, ('q7','q10'):60, ('q8','q10'): 0, ('q9','q10'):0,
             ('q0','q11'):32, ('q1','q11'):16, ('q2','q11'):- 32, ('q3','q11'):- 16.0,
             ('q4','q11'):-60, ('q5','q11'):- 30, ('q6','q11'):60, ('q7','q11'):30,
             ('q8','q11'):0, ('q9','q11'):0, ('q10','q11'):72}

Q = dict(linear)
Q.update(quadratic)



# Minor Embed and sample 1000 times on a default D-Wave sytem
sampleset = sampler_auto.sample_qubo(Q, num_reads=10000)
print(sampleset)
```
