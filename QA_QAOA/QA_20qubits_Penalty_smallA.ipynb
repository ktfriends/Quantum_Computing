```python
def QC_Computing(input:str):
    input = input.replace(' ','')
    input = input.replace('-','+-')
    list1 = input.split('+')

            
    dic1 = {}
    dic2 = {}
    for i in list1:
        if i.count('q')==1:
            tu = (i[i.index('q'):],i[i.index('q'):])  # i에서 q가 나오는 index
            dic1[tu] = float(i[:i.index('q')]) # int -> float 코드 수정(fractional은 계수가 분수라..)
        elif i.count('q')==2:
            tu = (i[i.index('q'):i.rfind('q')],i[i.rfind('q'):])
            dic2[tu] = float(i[:i.index('q')])

    return dic1, dic2

input = '6 + 24 q0 + 8 q1 + 16 q0 q1 + 12 q10 + 16 q0 q10 + 8 q1 q10 + 3 q11 + 8 q0 q11 + 4 q1 q11 + 12 q10 q11 + 8 q12 + 16 q0 q12 + 8 q1 q12 + 2 q13 + 8 q0 q13 + 4 q1 q13 + 8 q12 q13 + 8 q14 - 16 q0 q14 - 8 q1 q14 + 2 q15 - 8 q0 q15 - 4 q1 q15 + 8 q14 q15 + 84 q16 - 48 q0 q16 - 24 q1 q16 + 24 q10 q16 + 12 q11 q16 - 48 q12 q16 - 24 q13 q16 + 48 q14 q16 + 24 q15 q16 + 15 q17 - 24 q0 q17 - 12 q1 q17 + 12 q10 q17 + 6 q11 q17 - 24 q12 q17 - 12 q13 q17 + 24 q14 q17 + 12 q15 q17 + 108 q16 q17 + 132 q18 + 48 q0 q18 + 24 q1 q18 - 24 q10 q18 - 12 q11 q18 + 48 q12 q18 + 24 q13 q18 - 48 q14 q18 - 24 q15 q18 + 39 q19 + 24 q0 q19 + 12 q1 q19 - 12 q10 q19 - 6 q11 q19 + 24 q12 q19 + 12 q13 q19 - 24 q14 q19 - 12 q15 q19  + 108 q18 q19 + 8 q2 - 16 q10 q2 - 8 q11 q2 - 16 q12 q2 - 8 q13 q2 + 16 q14 q2 + 8 q15 q2 + 48 q16 q2 + 24 q17 q2 - 48 q18 q2 - 24 q19 q2 - 8 q10 q3 - 4 q11 q3 - 8 q12 q3 - 4 q13 q3 + 8 q14 q3 + 4 q15 q3 + 24 q16 q3 + 12 q17 q3 - 24 q18 q3 - 12 q19 q3 + 16 q2 q3 + 16 q4 - 32 q0 q4 - 16 q1 q4 - 16 q10 q4 - 8 q11 q4 + 48 q16 q4 + 24 q17 q4 - 48 q18 q4 - 24 q19 q4 + 32 q2 q4 + 16 q3 q4 - 4 q5 - 16 q0 q5 - 8 q1 q5 - 8 q10 q5 - 4 q11 q5 + 24 q16 q5 + 12 q17 q5 - 24 q18 q5 - 12 q19 q5 + 16 q2 q5 + 8 q3 q5 + 48 q4 q5 + 80 q6 + 32 q0 q6 + 16 q1 q6 + 16 q10 q6 + 8 q11 q6 - 48 q16 q6 - 24 q17 q6 + 48 q18 q6 + 24 q19 q6 - 32 q2 q6 - 16 q3 q6 + 28 q7 + 16 q0 q7 + 8 q1 q7 + 8 q10 q7 + 4 q11 q7 - 24 q16 q7 - 12 q17 q7 + 24 q18 q7 + 12 q19 q7 - 16 q2 q7 - 8 q3 q7 + 48 q6 q7 + 12 q8 - 16 q0 q8 - 8 q1 q8 - 24 q16 q8 - 12 q17 q8 + 24 q18 q8 + 12 q19 q8 + 16 q2 q8 + 8 q3 q8 + 16 q4 q8 + 8 q5 q8 - 16 q6 q8 - 8 q7 q8 + 3 q9 - 8 q0 q9 - 4 q1 q9 - 12 q16 q9 - 6 q17 q9 + 12 q18 q9 + 6 q19 q9 + 8 q2 q9 + 4 q3 q9 + 8 q4 q9 + 4 q5 q9 - 8 q6 q9 - 4 q7 q9 + 12 q8 q9'

linear, quadratic = QC_Computing(input)
```
