```python
def QC_Computing(input:str):
    input = input.replace(' ','')
    input = input.replace('-','+-')
    list1 = input.split('+')

            
    dic1 = {}
    dic2 = {}
    for i in list1:
        if i.count('q')==1:
            tu = (i[i.index('q'):],i[i.index('q'):])  # i에서 q가 나오는 index
            dic1[tu] = float(i[:i.index('q')]) # int -> float 코드 수정(fractional은 계수가 분수라..)
        elif i.count('q')==2:
            tu = (i[i.index('q'):i.rfind('q')],i[i.rfind('q'):])
            dic2[tu] = float(i[:i.index('q')])

    return dic1, dic2

input = '2102 + 760 q0 + 226 q1 + 616 q0 q1 - 700 q10 - 792 q0 q10 - 396 q1 q10 - 444 q11 - 396 q0 q11 - 198 q1 q11 + 376 q10 q11 + 2848 q12 - 64 q0 q12 - 32 q1 q12 - 504 q10 q12 - 252 q11 q12 + 1270 q13 - 32 q0 q13 - 16 q1 q13 - 252 q10 q13 - 126 q11 q13 + 616 q12 q13 - 1616 q14 + 64 q0 q14 + 32 q1 q14 + 504 q10 q14 + 252 q11 q14 - 1232 q12 q14 - 616 q13 q14 - 962 q15 + 32 q0 q15 + 16 q1 q15 + 252 q10 q15 + 126 q11 q15 - 616 q12 q15 - 308 q13 q15 + 616 q14 q15 + 472 q2 - 1232 q0 q2 - 616 q1 q2 + 792 q10 q2 + 396 q11 q2 + 64 q12 q2 + 32 q13 q2 - 64 q14 q2 - 32 q15 q2 + 82 q3 - 616 q0 q3 - 308 q1 q3 + 396 q10 q3 + 198 q11 q3 + 32 q12 q3 + 16 q13 q3 - 32 q14 q3 - 16 q15 q3 + 616 q2 q3 - 1152 q4 - 480 q0 q4 - 240 q1 q4 + 640 q10 q4 + 320 q11 q4 - 768 q12 q4 - 384 q13 q4 + 768 q14 q4 + 384 q15 q4 + 480 q2 q4 + 240 q3 q4 - 716 q5 - 240 q0 q5 - 120 q1 q5 + 320 q10 q5 + 160 q11 q5 - 384 q12 q5 - 192 q13 q5 + 384 q14 q5 + 192 q15 q5 + 240 q2 q5 + 120 q3 q5 + 560 q4 q5 + 2272 q6 + 480 q0 q6 + 240 q1 q6 - 640 q10 q6 - 320 q11 q6 + 768 q12 q6 + 384 q13 q6 - 768 q14 q6 - 384 q15 q6 - 480 q2 q6 - 240 q3 q6 - 1120 q4 q6 - 560 q5 q6 + 996 q7 + 240 q0 q7 + 120 q1 q7 - 320 q10 q7 - 160 q11 q7 + 384 q12 q7 + 192 q13 q7 - 384 q14 q7 - 192 q15 q7 - 240 q2 q7 - 120 q3 q7 - 560 q4 q7 - 280 q5 q7 + 560 q6 q7 + 1452 q8 + 792 q0 q8 + 396 q1 q8 - 752 q10 q8 - 376 q11 q8 + 504 q12 q8 + 252 q13 q8 - 504 q14 q8 - 252 q15 q8 - 792 q2 q8 - 396 q3 q8 - 640 q4 q8 - 320 q5 q8 + 640 q6 q8 + 320 q7 q8 + 632 q9 + 396 q0 q9 + 198 q1 q9 - 376 q10 q9 - 188 q11 q9 + 252 q12 q9 + 126 q13 q9 - 252 q14 q9 - 126 q15 q9 - 396 q2 q9 - 198 q3 q9 - 320 q4 q9 - 160 q5 q9 + 320 q6 q9 + 160 q7 q9 + 376 q8 q9'

linear, quadratic = QC_Computing(input)
```
